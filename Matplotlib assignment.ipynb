#### 1.2.Problem Statement
Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
# Fitting it to the periodic function
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
months = np.arange(12)
def tem_year(times,avg,ampl,time_offset):
    return (avg+ampl*np.cos((times+time_offset)*1.8*np.pi/times.max()))

res_max, cov_max= curve_fit(tem_year,months,Max,[40,20,0])
res_min, cov_min= curve_fit(tem_year,months,Min,[-40,20,0])

days=np.linspace(0,12,num=365)

plt.figure()
plt.plot(months,Max,'ro')
plt.plot(days,tem_year(days,*res_max),'r-')
plt.plot(months,Min,'bo')
plt.plot(days,tem_year(days,*res_min),'b-')
plt.xlabel("Month")
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

Charts to plot:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titan=pd.read_csv(data)    
```

```python
X=titan.sex.value_counts()
plt.pie(X,labels=['male','female'],autopct='%1.1f%%',radius=2)
plt.show()
```

```python
Male =titan[titan.sex=='male'].plot.scatter('age' , 'fare' ,c= 'blue' , label = 'male' )
titan[titan.sex=='female'].plot.scatter('age' , 'fare' ,c= 'orange' , label = 'female' , ax = Male)
plt.show()
```
